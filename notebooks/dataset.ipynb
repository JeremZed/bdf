## Nettoyage des données

```python
# Suppression des valeurs manquantes
d.drop_missing_values()
```

```python
# Remplacement des valeurs manquantes par la moyenne pour uniquement la colonne "stroke"
d.fill_missing(strategy='median', columns=['stroke'])
```

```python
# Suppression des lignes dupliquées
d.drop_duplicated_values()
```

```python
# Changement de type de la colonne price
d.convert_dtypes({'price' : float})
```

## Traitement sur les données

```python
# Normalisation des données
print("Avant", "min de price", d.df['price'].min(), "max de price", d.df['price'].max())
d.normalize(columns=['price'])
print("Après", "min de price", d.df['price'].min(), "max de price", d.df['price'].max())
```

```python
# Standardisation des données
print("Avant", "min de width", d.df['width'].min(), "max de width", d.df['width'].max())
d.normalize(columns=['width'])
print("Après", "min de width", d.df['width'].min(), "max de width", d.df['width'].max())
```
